## So What and why matplotlib

#### Matplotlib is a Python 2D plotting library that produces publication-quality figures in various formats and interactive environments. It can generate plots, histograms, bar charts, scatterplots, and more

```python
# Import necesery libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
from numpy import random
```

```python
# If you want to see the version 
print(matplotlib.__version__)
```

#### plt.plot() generates a line plot by default.
#### plt.show() is used to display the plot.
#### Use plt.legend() to display the legend
#### Use the label argument to create a legend
###### Properties of the plot() Function:
###### Color: The color of the line or marker can be set using the color argument or shortcuts.
###### Line Style: Control the style (solid, dashed, etc.) using the linestyle argument.
###### Markers: Control the appearance of points using marker.
###### Title, Labels: Add a title, x-labels, and y-labels for better understanding

### Line chart

```python

xp = np.array([1,2,3,4,5])
yp = np.array([2,34,5,6,8])

plt.plot(xp,yp)
plt.show()
```

#### For customization

```python
plt.plot(xp, yp, color='blue', linestyle='--', linewidth=2, marker='*')
plt.title('Line chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### FOr more decoration
###### color: Set the color of the line ('red', 'blue', '#FF5733' for hex values).
###### linestyle: Options include '-' for solid, '--' for dashed, and '-.' for dash-dot.
###### marker: Control the shape of points (e.g., 'o', '^', '*').
###### linewidth: Controls the width of the line

```python
# With line only points
xp = np.arange(1,6)
yp = np.random.randint(20,40, size=5)

plt.plot(xp,yp, "o")
plt.show()
```

```python
# look like Merging both chart line and scatter but it is still line chart
yp = random.randint(3,27, size=8)
plt.plot(yp, "o-.g")
plt.show()
```

```python
# Another Example
yp = np.random.rand(6)
plt.plot(yp, "o-.g")
plt.show()
```

```python
# Example with decoration and use of different properties
yp = np.random.randint(2,16, size=6)
plt.plot(yp, marker = "o",mfc = "y", ms = 15, mec = "r")
print(yp)
plt.show()
```

```python
xp = np.random.randint(2,16, size=6)
yp = np.random.randint(2,16, size=6)
plt.plot(xp, yp, marker = "o",mfc = "y", ms = 15, mec = "r")
print(yp)
plt.title("Random Data")
plt.xlabel("X-random")
plt.ylabel("Y-random")
plt.grid()
plt.show()
```

### Bar chart

```python
x = ["c++","java","python","Kotlin","ruby","C", "Node"]
y = np.random.randint(0,100, size=7)

plt.bar(x,y)
plt.xlabel("Languages")
plt.ylabel("Papularity")
plt.title("Programing Languages(2019,2022)",fontsize = 15)
plt.show()
```

```python
# Example
x = ["c++","java","python","Kotlin","ruby","C", "Node"]
y = np.random.randint(0,100, size=7)

plt.xlabel("Languages")
plt.ylabel("Papularity")
plt.title("Programing Languages(2019,2022)",fontsize = 15)
c = ['y','b','c','r','g','b','w']
plt.bar(x,y, color = c)
plt.show()
```

```python
# Example with more in depth use
x = ["c++","java","python","Kotlin","ruby"]
y = np.random.randint(0,100, size=5)
z = np.random.randint(1,90, size=5)

plt.xlabel("Languages")
plt.ylabel("Papularity")
plt.title("Programing Languages(2019,2022)",fontsize = 15)
c = ['y','b','c','r','g']
plt.bar(x,y, color = "y", label = "Previous", width=0.5)
plt.bar(x,z, color = "b", label="Now", width=0.3)
plt.legend()
plt.show()
```

```python
# Example
x = ["c++","java","python","Kotlin","ruby"]
y = np.random.randint(0,100, size=5)
z = np.random.randint(1,90, size=5)

width = 0.3
p = np.arange(len(x))
print(p)
p1 = [j+width for j in p]
print(p1)

plt.xlabel("Languages")
plt.ylabel("Papularity")
plt.title("Programing Languages(2019,2022)",fontsize = 15)
c = ['y','b','c','r','g']
plt.bar(p,y, width, color = "y", label = "Previous")
plt.bar(p1,z, width, color = "b", label="Now")
plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

### Scatter chart

```python
x = np.random.randint(0,100, size=5)
y = np.random.randint(1,90, size=5)
color = ['r','y','b','r','g']
size = [60,120,200,150,300]
plt.title("Scatter Plot", fontsize = 15)
plt.xlabel("Values")
plt.ylabel("Steps")
plt.scatter(x,y, size, color = c, marker = "*")
plt.show()
```

```python
# Two different dataset on same chart
x = np.random.randint(0,50, size=25)
y = np.random.randint(1,50, size=25)
color = ['r','y','b','r','g']
size = [60,120,200,150,300]
plt.title("Scatter Plot", fontsize = 15)
plt.xlabel("Values")
plt.ylabel("Steps")
plt.scatter(x,y,marker = "o", color = "black")
y = np.random.randint(1,50, size=25)
x = np.random.randint(0,50, size=25)
plt.scatter(x,y,marker = "o", color = "b")
plt.show()
```

### Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

##### bins: Controls the number of bins (intervals) for the histogram.
##### edgecolor: Controls the color of the bin edges

```python
n = np.random.randint(1,100, size=40)
plt.hist(n, width= 5)
plt.show()
```

### Pie Chart

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is a circle.
plt.show()
```

#### Automate as you want
##### autopct: Adds percentage labels to the chart ('%1.1f%%').
##### startangle: Rotates the start of the pie chart by the specified angle

## Subplots

##### subplots() creates multiple plots in a grid format (e.g., 2x2 grid)

```python
fig, axs = plt.subplots(2, 2)

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [20, 30, 35, 40]

# First subplot
axs[0, 0].plot(x, y1, 'r--')
axs[0, 0].set_title('Plot 1')

# Second subplot
axs[0, 1].plot(x, y2, 'g--')
axs[0, 1].set_title('Plot 2')

plt.tight_layout()
plt.show()
```

## Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, patch_artist=True, notch=True, boxprops=dict(facecolor='red'))
plt.title('Boxplot Example')
plt.show()


# Modify properties as per your need
# patch_artist: Colors the boxes.
# notch: Adds notches to the boxplots to compare medians
```

### 3D Plots (Requires mpl_toolkits module)

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()

# Modify properties as per your need
# 3D plot: plot_surface() creates 3D surface plots.
# cmap: Controls the colormap ('viridis', 'plasma', 'inferno')
```

### Real-World Example: Line Plot with Data

```python
gdp = [3000, 4000, 5000, 6000, 7000]
life_expectancy = [60, 65, 70, 75, 80]
population = [20, 50, 30, 10, 40]  # Population size for bubble size

plt.scatter(gdp, life_expectancy, s=population, alpha=0.5)
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```
